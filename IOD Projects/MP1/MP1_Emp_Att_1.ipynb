## Age vs. QuitStatus

The H0 (Null Hypothesis): There is no relationship between 'Age' and 'Quitting'.

The H1 (Alternative Hypothesis): There is a relationship between 'Age' and 'Quitting'.

We can reject the null hypothesis if the p-value < 0.05.

```python
from scipy import stats
```

```python
labels = ['Young', 'Old']

emp_att_dropped['Age_binned'] = pd.cut(emp_att_dropped['Age'], bins = 2, labels = labels)
emp_att_dropped.head()
```

```python
emp_att_dropped['Age_binned'].value_counts()
```

```python
AgeVSQuitStatus = emp_att_dropped.groupby(['Age_binned', 'QuitStatus']).size()
AgeVSQuitStatus
```

```python
AgeVSQuitStatus = pd.crosstab(emp_att_dropped['Age_binned'], emp_att_dropped['QuitStatus'])
AgeVSQuitStatus
```

```python
AgeVSQuitStatus.plot(kind = 'bar', stacked = True, rot = 0 )
plt.show()
```

```python
# Chi-square test BusinessValueVSQuitStatus
stat4, p4, dof4, expected4 = stats.chi2_contingency(AgeVSQuitStatus)
stat4, p4, dof4, expected4
```

```python
# interpret p-value
alpha = 0.05
print("p4 value is " + str(p4))
if p4 <= alpha:
    print('We reject our null hyphotesis (reject H0)')
else:
    print('We fail to reject our null hyphotesis (H0 holds true)')
```

```python
N = 100
young_worker = emp_att_dropped[emp_att_dropped['Age'] <= 33]['Salary'].sample(N)
old_worker = emp_att_dropped[emp_att_dropped['Age'] > 33]['Salary'].sample(N)
```

```python
young_worker
```

```python
old_worker
```

```python
ax = sns.distplot(old_worker, label='Old Worker', kde=False);
sns.distplot(young_worker, ax=ax, label='Young Work', kde=False);
plt.legend(loc = 'upper right')
plt.xlabel('')
plt.ylabel('')
plt.show();
```

H0: There is no difference between distribution of `Older Worker` & `Younger Worker`

H1: There is difference between distribution of `Older Worker` & `Younger Worker`

```python
t, p = stats.ttest_ind(old_worker, young_worker)
print("t = " + str(t))
print("p = " + str(p))
```

```python
alpha = 0.05
df = 2 * N - 2
p_value = 1 - stats.t.cdf(abs(t), df=df)  # #p-value after comparison with the t 
print("t = " + str(t))
print("p = " + str(2 * p))
```

```python
def test_result(p_value, alpha):
    
    if p_value < (alpha):
        print('Result of the hypothesis test: We reject the null hypothesis')
    else:
        print('Result of the hypothesis test: We failed to reject the null hypothesis')
```

```python
test_result(p_value, alpha)
print(test_result)
```
