```python
import numpy as np
import pandas as pd
import math

import statsmodels.api as sm
import statsmodels.tsa.api as smt
import statsmodels.formula.api as smf
from statsmodels.tsa.stattools import adfuller
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
%matplotlib inline

import itertools
import warnings
warnings.filterwarnings('ignore')
```

```python
filename = 'pune_1965_to_2002.csv'
rainfall_data_matrix = pd.read_csv(filename)
```

```python
rainfall_data_matrix.head()
```

```python
# Set 'Year' as index

rainfall_data_matrix.set_index('Year', inplace = True)
rainfall_data_matrix.head()
```

```python
# Transpose data for visualization

rainfall_data_matrix = rainfall_data_matrix.transpose()
rainfall_data_matrix
```

```python
# Genearete dates from 1965-01(January 1965) to 2002-12(December 2002)

dates = pd.date_range(start='1965-01', freq='MS', periods=len(rainfall_data_matrix.columns)*12)
dates
```

```python
# Visualize the whole data

plt.figure(figsize=(13,7))
plt.plot(rainfall_data_matrix)
plt.xlabel('Year')
plt.ylabel('Precipitation(mm)')
plt.title('Month vs Precipitation across all years')
```

```python
# Visualize with boxplot to understand the data better

plt.figure(figsize=(10,5))
plt.boxplot(rainfall_data_matrix)
plt.xlabel('Month')
plt.ylabel('Precipitation(mm)')
plt.title('Month vs Precipitation across all years')
```

```python
df = rainfall_data_matrix.transpose()
df.head(1)
```

```python
test = df.reset_index()
test.head()
```

```python
test = pd.melt(test, id_vars = 'Year', value_vars = test.columns[1:], var_name = 'month', value_name = 'rainfall')
test.head()
```

```python
monthly_rainfall = test.groupby(test['month']).rainfall.mean()
monthly_rainfall.plot()
plt.show()
```

```python
test.dtypes
```

```python
test.info()
```

```python
test['yr'] = test.Year.astype('string')
test['mth'] = test.month.astype('string')
```

```python
test.info()
```

```python
test.head(1)
```

```python
combined = test.yr.str.cat(test.mth, sep = ' ')
```

```python
test['year_and_month'] = pd.to_datetime(combined)
```

```python
test.info()
```

```python
test.head()
```

```python
test.set_index('year_and_month', inplace = True)
```
