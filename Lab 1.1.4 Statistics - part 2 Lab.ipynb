````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    '''
    #1
    
    X1 = np.array(X)
    Y1 = np.array(Y)
    
    #2
    x_sum = np.sum(X1)
    x_sqr_sum = np.sum(X1**2)
    y_sum = np.sum(Y1)
    xy_sum = np.sum(X1*Y1)
    
    #3
    
    m= ((len(X1) * xy_sum) - (x_sum*y_sum)) / ((len(X1) * x_sqr_sum) - (x_sum ** 2))
    
    #4
    
    b = (y_sum - (m * x_sum)) / len(X1)
    
    return m, b
    
   
    
    m = 1
    b = 1
    
   

# solution
m, b = best_fit(X, Y)
#best fit line:
#y = 0.80 + 0.92x

#5
    
y = (m * np.array(X)) + b

# plot points and fit line
# ANSWER (~3 lines)

plt.scatter(X, Y)
plt.plot(X, y, color = 'red')
plt.show()
````

# 11. [BONUS] Calculate Mean Confidence Interval

Check following methods:

1. scipy.stats.sem
2. scipy.stats.t.ppf

```
# Expected Output:
(30.476190476190474, 19.539561081583383, 41.41281987079756)
```

````python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    a= 1.0*np.array(data)
    n= len(a)
    m, se = np.mean(a), scipy.stats.sem(a)
    h = se * scipy.stats.t.ppf((1+confidence) / 2., n-1)
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))
````
